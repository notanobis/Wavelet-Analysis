## Build a synthetic signal
$$y_0(t) = A_0 sin(2 \pi f_0t)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dt = 1
N = 10**3
T = N*Dt 
n = 100
i = np.arange(N)
t = Dt*i
A0 = 2
f0 = n/T 
y0 = A0 * np.sin(2*np.pi*f0*t)
plt.plot(y0)
```

```python
fmin = 1/n
fmax = N/(2*T)
fmax
```

```python
u = np.zeros(N,dtype=complex)
n = np.arange(N)

for k in n:
    for l in i:
        u[k] += 1/N * (y0[l] * np.exp(-2*np.pi*1j*k*l/N))
```

```python
u2 = u[:int(N/2)]
S = 2*T* np.square(np.abs(u2))/N**2
plt.plot(S)
```
